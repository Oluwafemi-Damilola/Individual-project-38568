```python
years= [2011, 2012, 2013, 2014, 2015, 2016, 2017, 2018, 2019, 2020]
duration= [103, 101, 99, 100, 100, 95, 96, 93, 90]

movie_dict= {year: dur for year, dur in zip(years, duration) }

print(movie_dict)
```

```python
import pandas as pd


durations_df = pd.DataFrame({'duration': movie_dict.values()}, index=movie_dict.keys())

print(durations_df)
```

```python
import matplotlib.pyplot as plt

# plotting

durations_df.plot(kind='line', marker=0)

plt.xlabel('years')
plt.ylabel('durations')
plt.title('Netflix Movie Durations 2011-2020')
plt.show()
```

```python
path = r"C:\Users\Damilola Oluwafemi\Desktop\class project\datasets\netflix_data.csv"

netflix_df = pd.read_csv(path)

print(netflix_df.head(5))
```

```python
netflix_df_movies_only = netflix_df.loc[netflix_df['type'].isin(['Movie'])]

netflix_movie_col_subset = netflix_df_movies_only[['title', 'country', 'genre', 'release_year', 'duration']]

print(netflix_movie_col_subset.head(5))
```

```python
import matplotlib.pyplot as plt
netflix_movie_col_subset.plot.scatter(x='release_year', y='duration')

plt.xlabel('release_year')
plt.ylabel('duration')
plt.title('Movie Duration By Year of Release')

plt.show()
```

movie duration has decreased over the years

```python
short_movies = netflix_movie_col_subset.loc[netflix_movie_col_subset['duration'] < 60]

print(short_movies.head(20))
```

```python
color = []
for index, row in netflix_movie_col_subset.iterrows():
    if row['genre'] == 'Children' :
        color.append('red')
    elif row['genre'] == 'Documentary' :
         color.append('blue')
    elif row['genre'] == 'Stand-Up' :
        color.append('green')
    else:
        color.append('black')

print(color[:10])
```

```python
netflix_movie_col_subset.plot.scatter(x='release_year', y= 'duration', c=color, alpha=0.8)
plt.xlabel('Release_year')
plt.xlabel('Duration(mins')
plt.title('Movie Duration by year of release')

plt.show()
```

From the analysis perfomed, movie duration made over the years seem to have declined.
